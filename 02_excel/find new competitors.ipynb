```python
# 读取正常竞争者清单
competitors_o = pd.read_excel(file_path,sheet_name='正厂竞争者数据汇总')
# competitors_o = dt[['Brand','Competitors']]
competitors_o['IsOEM'] = '1'

# 读取副厂竞争者清单
competitors_a = pd.read_excel(file_path,sheet_name='副厂竞争者数据汇总')
# competitors_a = dt[['Brand','Competitors']]
competitors_a['IsOEM'] = '0'

competitors = pd.concat([competitors_o[['Brand','Competitors','IsOEM']],competitors_a[['Brand','Competitors','IsOEM']]],axis=0).reset_index()
```

```python
def domain_edit(competitors):
    if len(str(competitors).split('.'))>2:
        return '.'.join(str(competitors).split('.')[-2:])
    else:
        return competitors
competitors['domain_check'] = competitors['Competitors'].map(domain_edit)
competitors.head(20)
```

```python
# 合并所有 comlevel 大于 0.1 的竞品
coml_path = r'F:\HolfGit\HolfCamus\02_excel\comlevel'
# path_group 简称 pg
pg = []
for p in os.listdir(coml_path):
    pg.append(os.path.join(coml_path,p))
```

```python
rs = []
for i in pg:
    tp = pd.read_excel(i,sheet_name='Sheet 2')
    # display(i)
    tp['Brand'] = i.split('\\')[-1].split('.')[0]
    rs.append(tp)
```

```python
# 新的竞品列表
new_comlevel_l = pd.concat(rs,axis=0)
new_comlevel_l.head()
```

```python
new_comlevel_lt = new_comlevel_l.loc[:,['Domain','Brand']]
new_comlevel_lt.head()
```

```python
res = competitors.merge(new_comlevel_lt,how='right',left_on=['Brand','domain_check'],right_on=['Brand','Domain'])
```

```python
new_competitors = res.loc[res['domain_check']!=res['Domain']]
new_competitors[['Brand','Domain']]
```
